```python
import pandas as pd
import numpy as np
import re
```

```python
df2018 = pd.read_csv('ExpedientesPublicados_2018.csv', sep=',', encoding='utf-8')
df2019 = pd.read_csv('ExpedientesPublicados_2019.csv', sep=',', encoding='utf-8')
df2020 = pd.read_csv('ExpedientesPublicados_2020.csv', sep=',', encoding='utf-8')
df2021 = pd.read_csv('ExpedientesPublicados_2021.csv', sep=',', encoding='utf-8')
df2022 = pd.read_csv('ExpedientesPublicados_2022.csv', sep=',', encoding='utf-8')
```

```python
#Funcion para obtener solamente los codigos cucops, si hay varios separarlos por comas
def getCucops(text):
  if text==text:
    cucops = re.findall(r'\d{4,8}', text)
    return ';'.join(cucops)
  return text

#Funcion para limpiar espacios y saltos de líneas
def deleteSpaces(text):
  if text==text:
    return re.sub(r'\s+', " ", text)
  return text
```

```python
print(getCucops('6190 y 2 claves 3 cucop 4 adicionales'))
print(getCucops('5690 '))
print(getCucops('2110, 2120, 2720, 12345678, '))
print(deleteSpaces('2110, \n2120,    2720, 12345678, '))
```
