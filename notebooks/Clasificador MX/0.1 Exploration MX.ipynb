```python
cucops= pd.read_csv('/content/drive/MyDrive/TFM/data/export_cucops.csv', sep=',', encoding='UTF8')
cucops.head()
```

```python
#CAPITULO
#¿Cuántos capitulos hay?
cucops.groupby(['id_capitulo'])['id_capitulo'].unique().count()
```

```python
#¿Cuantas partidas genericas hay por concepto?
cucops.groupby(['id_capitulo', 'id_concepto', 'id_partida_generica']).size().reset_index(name='count').groupby(['id_capitulo','id_concepto']).count()['count'].plot(kind='bar', figsize=(30,7))
```
